```python
import pandas as pd

data = pd.read_csv('results.csv')
info = pd.read_csv('GenericQuestionnaire.csv')
participants = data.columns[4:]

data_entropy = pd.read_csv('entropy.csv')

data = pd.read_csv('results.csv')
cols = data.columns.tolist()
with open('dataset.csv', 'w') as output_file:
    import csv
    writer = csv.writer(output_file)
    headers = ['Participant'] + cols[:4] + ['Entropy', 'Sequence'] + ['Answer'] + ['Q%d' % d for d in range(1, 16)]
    writer.writerow(headers)
    for row in data.itertuples():
        row = list(row)[1:]
        for participant in participants:
            participant_index = cols.index(participant)
            entropy = ''
            row_entropy = data_entropy[(data_entropy['Block'] == row[0]) & (data_entropy['Song'] == row[1]) & (data_entropy['Real'] == row[2]) & (data_entropy['Shuffle'] == row[3])]
            print(row[:4])
            entropy = row_entropy['Entropy'].tolist()[0]
            sequence = row_entropy['Sequence'].tolist()[0]
            new_row = [participant] + row[:4] + [entropy, sequence] + [row[participant_index]]
            for inforow in info[info['Participant'] == participant].itertuples():
                new_row += list(inforow)[2:]
                break
            writer.writerow(new_row)
```
